### Check for Stationarity

```python
import numpy as np 
import pandas as pd
```

```python
# Read the daily website visits data
visits=pd.read_csv('website-visits.csv')
```

```python
# Data Cleaning
visits=visits.replace(',','',regex=True)
visits['PageLoads']=visits['PageLoads'].astype(int)
visits['UniqueVisits']=visits['UniqueVisits'].astype(int)
visits['FirstTimeVisits']=visits['FirstTimeVisits'].astype(int)
visits['ReturningVisits']=visits['ReturningVisits'].astype(int)

#Convert to datetime
visits['Date']= pd.to_datetime(visits['Date'])

#Check
visits.head()
```

```python
import matplotlib.pyplot as plt
import plotly.express as px

px.line(visits, x='Date', y='UniqueVisits',
                title='Is the data Stationary?')
```

```python
#Stationarity Check - Augmented Dickey-Fuller Test
from statsmodels.tsa.stattools import adfuller

adftest = adfuller(visits['UniqueVisits'], autolag='AIC')
dfoutput=pd.Series(adftest[0:4],
                   index=['Test Statistic','p-value','#lags','#observations'])

for key,value in adftest[4].items():
    dfoutput['critical value (%s)'%key]= value
print(dfoutput)
```

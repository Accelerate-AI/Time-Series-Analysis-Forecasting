```python
#Q3: Find the correlation between FAANG stocks and  NASDAQ on a 6 month moving window.
```

```python
import pandas as pd
import pandas_datareader as pdr               #for retreiving stock prices

start = pd.Timestamp('2015-1-1')
end = pd.Timestamp('2020-12-31')

#Nasdaq Index daily data
nasdaq = pdr.get_data_yahoo('ndaq',start, end)['Adj Close']

#FAANG Stocks daily close data
tickers = ['meta', 'aapl', 'amzn', 'nflx', 'goog']
faang_df = pdr.get_data_yahoo(tickers, start, end)['Adj Close']
```

```python
ndaq_returns = nasdaq.pct_change()
faang_returns = faang_df.pct_change()
```

```python
corr = faang_returns.rolling(180).corr(ndaq_returns)            #180 days MA 
corr.plot()
```

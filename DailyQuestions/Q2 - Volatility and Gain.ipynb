```python
# Q2. Consider the Ford stock price data from Yahoo Finance.
import pandas as pd
from pandas_datareader import data
start = pd.Timestamp('2017-1-1')
end = pd.Timestamp('2021-12-31')
f = data.DataReader("F", 'yahoo', start, end)

# 1. Is the variance of the trading volume relatively stable over time?
# 2. On how many days did the stock close higher than it opened?
```

#### Solution:

#### 1. Is the variance of the trading volume relatively stable over time?

```python
# We can look at the plot of standard deviation at a monthly frequency
f.resample("M").std().plot()
```

```python
# Another way it to view the box-plot of ts, at at monthly frequency.
import seaborn as sns
import matplotlib.pyplot as plt

f['Month-Year'] = f.index.to_period('M')

plt.figure(figsize=(10,6))
sns.boxplot(data=f, x='Month-Year', y='Volume')
plt.xticks(rotation = 'vertical')
plt.show()
```

#### 2. On how many days did the stock close higher than it opened?

```python
# Stock gained if Close > Open  
f['gain'] = f.Close > f.Open
f.gain.sum()
```

```python
# The above dosen't tell us much. 

# Let's look at the no of says the stock gained each month.
f.resample("M").sum().gain.plot()
```
